## Analitik Nedir? ##
Analitik, Parça ile bütün arasındaki ilişkiyi çözmek ve özüne inmektir. Diğer anlamı ise verinin içindeki anlamlı deseni keşfetme sanatıdır.

İş Analitiği, veri yönetimi ve ticari bilgi gibi temel konuların yanı sıra tahminsel modelleme, optimizasyon ve simülasyon gibi sofistike konuları kapsamaktadır. 


### Analitik Katmanlar ###
__Tanımlayıcı Analitik (Descriptive):__ Olanları daha iyi anlamak, ilgili veriyi bulmak ve ilginç örüntüleri tanımlamak için yaratıcı teknikleri kullanmaktır. En önemli teknik Veri Madenciliği ’dir .

__Kestirimci Analitik (Predictive):__ Verileri kullanarak ileride ne olacağını öngörmektir. İstatistiksel tahmin yöntemleri ve benzetim uygun yöntemlerdir

__Kural Koyucu Analitik (Prescriptive):__ Veriyi kullanarak gelecekte ne ve nasıl yapılması gerektiğini belirlemektir. YA’nın optimizasyon teknikleri genellikle bu kategoride kullanılır

## Yöneylem Araştırması (YA) ##
İngilizcesi Operations Research (OR)'tür.

__Operation (Operasyon):__ Bir organizasyonda yapılan tüm faaliyetlerdir<br>
__Research (Araştırma):__ Bilimsel metotlarla gözlem ve test yapmaktır

YA, karar verme problemlerine bilimsel metotları kullanarak sayısal çözüm yaklaşımıdır

### Yöneylem Araştırması Nedir? ###
Dünyanın gerçek karmaşık sorunları için matematiksel modelleme, istatistik ve algoritma gibi bilimsel yöntemleri kullanan disiplinlerarası bir bilimdir. 

İnsan, makina, para ve malzemeden oluşan endüstriyel, ticari, resmi ve askeri sistemlerde yönetimlerde karşılaşılan problemlere bilimsel yaklaşımdır. 

Mevcut imkanlardan en büyük faydayı sağlamak için girişilen bilimsel yaklaşımlar ve teknikler cümlesidir

Herhangi bir sistemin karmaşık operasyonlarını anlamak ve performansını iyileştirmek için bilimsel karar verme yöntemleridir

### Yöneylem Araştırmasının Temel Özellikleri! ###
__Sistem yaklaşımını kullanması:__ Yöneylem araştırması problemi çözerken, o problemin ait olduğu organizasyonun bütün unsurlarını, çevresini ve aralarındaki etkileşimi göz önünde bulundurur

__Disiplinler arası bir yaklaşım olması:__ Problemin modellenmesinde ve çözümünde farklı bakış açılarından faydalanabilmek için problemlerin disiplinler arası bir ekip tarafından incelenmesi gerekir

__Bilimsel yöntemler kullanması:__ Yöneylem araştırması problemi tanımlar, modeller, çözer, sonuçları test eder ve uygular

__Bilgisayar teknolojisine güvenmesi:__ Problemlerin çözülebilmesi için bilgisayar teknolojisine gereksinim duyar.


## Optimizasyon Örnekleri ##
Bir problemin optimizasyon problemi olması için 2 temel ihtiyaç vardır. 

1- Kaynakların kısıtlı olması gerekmektedir.<br>
2- Seçim yapmamızı gerektiren alternatiflerin olması gerekmektedir.

Bu iki durumun olduğu herhangi bir problem bizim için optimizasyon problemidir. Optimizasyonu anlamak için en basit problem __Sırt Çantası Problemidir.__ Gelin şidmi kısaca bu problemi inceleyelim.


### Sırt Çantası Problemi ###
Bir hırsızın 1 kg kapasiteli bir çanta ile bir eve hırsızlık için girdiğini düşünelim. Hacim olarak istediği büyüklükte ürün çalabilse de, ürünlerin toplam ağırlıkları 1 kg geçemez. Bu hırsız eve girdiğinde 5 farklı ürünün yükte hafif paha da ağır olduğunu farkediyor. İnternette yaptığı araştırma sonucunda da bu ürünlerin ağırlık ve fiyatlarının şu şekilde olduğunu görüyor.



Hırsızımız aç gözlü (greedy) bir insan olduğu için aklında şöyle bir algoritma geliştiriyor. Ben en pahalı üründen itibaren çantamı doldurabilirsem, günün sonunda en pahalı ürünleri çalabilirim. Bu algoritmaya göre çalmaya başlıyor. Her çaldığı üründen sonra kapasitesini tekrar hesaplayarak çalıyor.



Günün sonunda 950TL'lik vurgun ile evi terkediyor. Peki, bu çözüm en iyi çözüm müdür? Eğer hırsızımız yöneylem araştırması tekniklerini bilseydi daha iyi bir hırsızlık yapabilir miydi?

Ürün sayımız az olduğu için tüm kombinasyonları denersek daha iyi bir çözüm olduğunu görebiliriz. Ürün sayımız fazla olsaydı da yöneylem araştırması tekniklerini kullanarak en iyi çözüme çok kısa bir sürede ulaşabilirdik. Hırsızımız da 50TL fazla para kazanarak, belki evine yürüyerek değilde taksi ile gidebilirdi.

## Yöneylem Araştırması Uygulama Adımları ##
<u><b>1- Problemin Tanımı</b></u><br>
__Karar;__  Mümkün seçenekler arasından bir faaliyet veya faaliyetler dizisinin benimsenmesine denir<br>
__Karar verici;__ alternatifler arasından en uygun olanını seçme konusunda karar verme yetkisine sahip birey ya da topluluğa verilen genel isimdir<br>

__Karar vericinin;__<br>
- Ulaşmak istediği bir amacı, <br>
- Bu amaca ulaşmada farklı alternatiflerin/seçeneklerin bulunması<br>
- Alternatifler içinden hangisinin amaca en uygun olduğu konusunda kuşkusunun olması

Ancak bu koşullarda bir problem vardır denir

<u><b>2- Sistemin Gözlemlenmesi ve Veri Toplama</b></u><br>
Sistem gözlemleme;
- Probleme etki eden parametreler tahmin edilmeye/ belirlenmeye çalışılır<br>
- Parametrelere ait veriler derlenir,<br>
- Tahmini değerler sabit sayılar olarak işleme tabi tutulurlar<br>

Problem elemanlarının duruma en uygun biçimde belirlenebilmesi için sistem yaklaşımı kullanılır

<u><b>3- Matematiksel Model Oluşturma</b></u><br>
Problemin kolayca çözülebilecek bir yapıya oturtulması gerekmektedir. Yani basitleştirilmelidir.<br>
Bu basitleştirmede genellikle MODELLEME olarak bilinir. <br>

Model nedir?<br>
- Bir şeyin küçültülmüş örneğidir. 
- Yapılacak bir şeyin şablonudur. 
- Bir şeyi görselleştirmeye yarayan tanımlamadır.
- Göz önüne alınan sistemi ifade etmekte ve sistem içindeki ilişkileri tanımlamaktır.


<u><b>4- Matematiksel Modelin Çözümü</b></u><br>
__Analitik çözüm:__
- Analitik çözümde sadece matematiğin değil iktisat teorisinin de temel kuralları kullanılır

__Algoritma çözümü:__
- Analitik çözüm bazen çok zor veya çözümsüz olabilir. 
- Belirli bir sıra içerisinde gerçekleştirilen matematiksel ve mantıksal işlemler kümesine “algoritma” denir. 
- Yinelemeli olarak uygulanan algoritmalar her adımda optimuma daha yakın bir çözüme doğru ilerler

__Benzetim/Simulasyonçözümü:__
- Problem, analitik olarak veya algoritmalarla çözülemiyorsa kullanılır. Sistemin davranış şekli bilgisayar ortamında taklit edilir

__Sezgisel(Heuristic) çözüm:__
- Problem optimum çözümü bulunamayacak kadar karmaşıksa, sezgisel yöntemler sezgiye veya bazı deneysel kayıtlara dayanan karar kuralları ile belirli sayıda adımdan sonra en iyi olmasa da tatminkar bir sonuç verirler

<u><b>5- Modelin Teyit Edilmesi</b></u><br>
Modelden elde edilen çözümü uygulamaya koymadan önce gerçeğe uygunluğunun kanıtlanması gerekir<br>
Eğer çözüm sistemin geçmiş dönem sonuçlarını aynen veya daha olumlu bir şekilde sağlıyorsa, modelin geçerli olduğu kabul edilir<br>
Eğer sistemin geçmiş dönem sonuçları yoksa benzetimden yararlanılır<br>
Model geçerliliğinin kanıtlanmasında bir başka yol olarak da sistemdeki deneyimli kişilerin görüşlerine başvurulabilir

<u><b>6- Çözümün Uygulaması ve Yorumu</b></u><br>
Yöneylem araştırması ekibi, uygulama sürecini açıklamalı ve uygulamada yardımcı olmalıdır<br>
Uygulamanın nasıl yapılacağı bir rapor halinde yönetime sunulmalıdır<br>
Modeli uygulamak için iyi dokümante edilmiş bir sistem kurulmalı<br>

## Doğrusal Programlama Nedir? ##
Doğrusal Programlama (DP), Lineer Programlama (LP) adı ile de anılır.

Doğrusal Programlama belli bir amacı gercekleştirmek icin sınırlı kaynakların etkin kullanımını ve ceşitli secenekler arasında en uygun dağılımını sağlayan matematiksel bir tekniktir.

DP deterministik bir aractır, yani model parametreleri belirgin olarak kabul edilir. 

Doğrusal Programlama kavramında bulunan “doğrusal” kelimesi ile anlatılmak istenen duşunce, girdiyi oluşturan değişkenler ile cıktı değeri arasında doğrusal bir ilişki olmasıdır. “Programlama” sozcuğu ile anlatılmak istenen ise elde bulunan kısıt ve imkanlar ile mumkun olan en uygun dağılım sonucu en yuksek karı elde edilen durumu bulmaya calışmaktır.


### Doğrusal programlama Tekniği Varsayımları ###
- Amac fonksiyonu ve kısıtlayıcı şartlar doğru tanımlanmalıdır. 
- Maksimizasyonu mu yoksa maliyet minimizasyonu mu olduğu acıkca belirtilmelidir.
- Değişkenler kantitatif olmalıdır. Doğrusal programlama kalitatif (rakamla ifade edilemeyen) değişkenler icin kullanılmaz.
- Değişkenler kendi aralarında ilişkili olmalıdır.
- Kullanılacak kaynaklar sınırlı olmalıdır.
- Değişkenler arasında kurulan bağıntılar doğrusal olmalıdır.
- Değişkenler arasında alternatif secim olanağı olmalıdır.
- Doğrusal programlamanın uygulanacağı işletme problemi kısa donemli olmalıdır.
- Karar değişkenlerinin sıfır ya da pozitif olması gerekir.

### Doğrusal Programlama Probleminin Matematiksel Modeli ###
Doğrusal Programlama Problemi (DPP) matematiksel olarak modellenirken, üç önemli kısma sahiptir.
- Amaç fonksiyonu
- Karar değikeni
- Kısıtlar (koşullar)


<b><u>Amaç fonksiyonu</u></b><br>
Amaç fonksiyonu, doğrusal programlama kapsamına giren problemdeki kısıtların kullanılmasıyla oluşan faydanın maksimize edilmesi veya yine kısıtların kullanılmasıyla oluşan zararın minimize edilmesi olarak tanımlanabilir. 

<b><u>Karar değikeni</u></b><br>
Bir karar modelinin çözümlenmesi sürecinde değeri hesaplanacak olan karar unsurlarıdır.

<b><u>Kısıtlar</u></b><br>
Modeli oluşturan kısıtların belirli bir sınırı olduğunu (kapasite), bu sınırın elindeki kısıt miktarlarına ait maksimum ya da minumum değer (stok, kapasite, vb.) olarak düşünüldüğü sınırlayıcı niteliğe sahip ifadedir.
